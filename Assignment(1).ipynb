# Mean of each column

```python
sl_mean = table1['SL'].mean()
print('Mean SL = ',sl_mean)
sw_mean = table1['SW'].mean()
print('Mean SW = ',sw_mean)
pl_mean = table1['PL'].mean()
print('Mean PL = ',pl_mean)
pw_mean = table1['PW'].mean()
print('Mean PW = ',pw_mean)
```

```python
plt.figure(figsize=(8,8))

plt.subplot(2,1,1)
plt.scatter(table2['SL'], table2['SW'], s=12, c='blue', marker = ".")
plt.title('Scatter Plot  SL v/s SW', fontsize = 10)
plt.xlabel('SL', fontsize = 10)
plt.ylabel('SW', fontsize = 10)

plt.subplot(2,1,2)
plt.scatter(table2['PL'], table2['PW'], s=12, c='red', marker = ".")
plt.title('Scatter Plot of PL v/s Pw', fontsize = 10)
plt.xlabel('PL',fontsize = 10)
plt.ylabel('PW', fontsize = 10)

plt.tight_layout()
```
