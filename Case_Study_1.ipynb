**IMPORT DATA**

The data set shows information about the features of various car models of different companies and also information relating to their sales.

```python
data = pd.read_csv('/content/cars_data.csv')
table = pd.DataFrame(data)
table
```

**Bar Graph for Male and Female buyers**

We can represent the number of female and male buyers who participated in the sales using a count plot. Countplot() is a function available by default in the seaborn librabry. It displays the number or count of elements present in a particular data set or a particular column of a data set.

Here, we are taking the count of the elements "female" and "male" from the column titled "Buyer Gender". 

```python
plt.figure(figsize=(12,8))
graph = sns.countplot(x='Buyer Gender',data=table,)
graph.set(ylim=(0, 6000))
plt.show()
```

From the figure, it is clear that almost an equal number of male and female buyers have participated in the sales.

**Top 5 Cars based on Sale Price**

To find the top 5 cars based on their respective sale price, we can use the simple sort_values() function.

We will first extract the "Sale Price" column along with "Make" and "Model" columns to help us identify the car. We can put these three columns into a tabular form and then use sort_values() function on the sale price column to arrange it in descending order. The other two columns will be arranged accordingly.

We can then display the first five rows of the arranged table, and it shows us the top 5 cars based on sale price.


```python
sell_price = table[["Make","Model","Sale Price"]]
top_five = sell_price.sort_values(by='Sale Price',ascending=False)
top_five.head(5)
```

The table shows the model and make of the top 5 cars based on their sale price.

**Least 5 Cars based on Resell Price**

To find the least 5 cars based on their resell price, we use the same process as before. We extract the columns "Make", "Model" and "Resell Price" and use the sort_values() function to sort the resell prices, this time in ascending order.

Then, if we display the first 5 rows of the arranged table, we get the desired result.

```python
resell_price = table[["Make","Model","Resell Price"]]
least_five = resell_price.sort_values(by='Resell Price',ascending=True)
least_five.head(5)
```

The table shows the model and make of the least 5 cars based on their resell price.
